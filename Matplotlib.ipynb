# Matplotlib

### Create the first visualization using 'matplolib.pyplot'

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 20000)
print(x)
```

```python
# Quadratic curve
y = x**2
print(y)
```

```python
# Line plot --> plt.plot(x, y)
plt.plot(x,y)



```
