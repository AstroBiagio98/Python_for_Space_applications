# Matplotlib

### Create the first visualization using 'matplolib.pyplot'

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 20000)
print(x)
```

```python
# Quadratic curve
ax = plt.figure().add_subplot(projection = '3d')
y = np.linspace(-100,100,20000) 
z = x**2 + y**2
print(y)
ax.plot(x,y,z)
plt.show()
```

```python
# Line plot --> plt.plot(x, y)



plt.plot(x,y)
plt.show()


```

#### The more the values in 'x' and 'y' the smoother the curve gets

scatter plot -> grafico di punti

## Customize the plot

```python
plt.figure(figsize= (4,4))
plt.plot(x,y)
plt.grid()
plt.title('Visualizing the quadratic data', color = 'm', weight = 'bold', fontsize = 12)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
